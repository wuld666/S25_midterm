The data in the mushrooms dataset is currently encoded with strings. We need to convert the categorical variables into numeric indicators. One way to achieve this is by using the pd.get_dummies function, which will create indicator variables for each category in the data, effectively converting them to a numeric representation that can be used by the algorithms in sklearn.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

mush = pd.read_csv('mushrooms.csv')
cat_mush = pd.get_dummies(mush)

# Subset data into X and y
X_mush = cat_mush.iloc[:,2:]
y_mush = cat_mush.iloc[:,1]
```

```python
# The following code reduces the data to two dimensions and plot.

# import modules
from sklearn.decomposition import PCA

# Create and fit PCA model
pca = PCA(n_components=2)
transformed = pca.fit_transform(X_mush)
xs = transformed[:, 0]
ys = transformed[:, 1]
df_trans = pd.DataFrame({'xs': xs, 'ys': ys})

# Create scatter plot
plt.scatter(df_trans.loc[y_mush == 0, 'xs'], df_trans.loc[y_mush == 0, 'ys'], c='tab:green', label='p')
plt.scatter(df_trans.loc[y_mush == 1, 'xs'], df_trans.loc[y_mush == 1, 'ys'], c='tab:blue', label='e')
plt.legend(loc='lower left')

# Show plot
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Mushroom Dataset')
```
